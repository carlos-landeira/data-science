```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# reads the dataset
file_path = "data/Automobile.csv"
df = pd.read_csv(file_path)

# shows dataset structure info
print("\nDataset Info:")
print(df.info())

# shows number of duplicate rows (full row comparision)
print("\nNumber of duplicate rows:")
print(df.duplicated().sum())

# removes duplicate rows, if any
df = df.drop_duplicates()
print("\nDuplicates removed. New shape of the dataset:")
print(df.shape)

# shows data missing for each of the variables/features
print("\nMissing data per feature:")
print(df.isnull().sum())

# treats missing data by filling it with the mean value of the respective feature/variable
df.fillna(df.mean(numeric_only=True), inplace=True)

# shows data missing for each of the variables/features after treatment
print("\nMissing data after treatment:")
print(df.isnull().sum())

# unique case, 1 one row has only 'subaru' (manufacturer name) in its name, i'll append the name with the model
df.loc[df['name'] == 'subaru', 'name'] = 'subaru dl'

# unique case 2, one row has only 'capri ii' (model name), i'll prepend the manufacturer name
df.loc[df['name'] == 'capri ii', 'name'] = 'ford capri ii'

# split 'name' column (which has manufacturer + model) into two separate columns: 'manufacturer' and 'name'
split_names = df['name'].astype(str).str.split(' ', n=1, expand=True)
df.insert(loc=0, column="manufacturer", value=split_names[0])
df['name'] = split_names[1].fillna('')

# correct manufacturer names and use standard name (like 'vw' to 'volkswagen')
df.replace({'manufacturer': {
    'vw': 'volkswagen',
    'vokswagen': 'volkswagen',
    'chevy': 'chevrolet',
    'chevroelt': 'chevrolet',
    'maxda': 'mazda',
    'mercedes': 'mercedes-benz',
    'toyouta': 'toyota',
    'hi': 'international harvester',
}}, inplace=True)

# count unique manufacturers
print("\nUnique manufacturers and their counts:")
print(df['manufacturer'].value_counts())

# shows the updated dataset
df
```
