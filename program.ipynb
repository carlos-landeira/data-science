```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import itertools

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# reads the dataset
file_path = "data/Automobile.csv"
df = pd.read_csv(file_path)

# shows dataset structure info
print("\nDataset Info:")
print(df.info())

# shows number of duplicate rows (full row comparision)
print("\nNumber of duplicate rows:")
print(df.duplicated().sum())

# removes duplicate rows, if any
df = df.drop_duplicates()
print("\nDuplicates removed. New shape of the dataset:")
print(df.shape)

# shows data missing for each of the variables/features
print("\nMissing data per feature:")
print(df.isnull().sum())

# treats missing data by filling it with the mean value of the respective feature/variable
df.fillna(df.mean(numeric_only=True), inplace=True)

# shows data missing for each of the variables/features after treatment
print("\nMissing data after treatment:")
print(df.isnull().sum())

# unique case, 1 one row has only 'subaru' (manufacturer name) in its name, i'll append the name with the model
df.loc[df['name'] == 'subaru', 'name'] = 'subaru dl'

# unique case 2, one row has only 'capri ii' (model name), i'll prepend the manufacturer name
df.loc[df['name'] == 'capri ii', 'name'] = 'ford capri ii'

# split 'name' column (which has manufacturer + model) into two separate columns: 'manufacturer' and 'name'
split_names = df['name'].astype(str).str.split(' ', n=1, expand=True)
df.insert(loc=0, column="manufacturer", value=split_names[0])
df['name'] = split_names[1].fillna('')

# correct manufacturer names and use standard name (like 'vw' to 'volkswagen')
df.replace({'manufacturer': {
    'vw': 'volkswagen',
    'vokswagen': 'volkswagen',
    'chevy': 'chevrolet',
    'chevroelt': 'chevrolet',
    'maxda': 'mazda',
    'mercedes': 'mercedes-benz',
    'toyouta': 'toyota',
    'hi': 'international harvester',
}}, inplace=True)

# count unique manufacturers
print("\nUnique manufacturers and their counts:")
print(df['manufacturer'].value_counts())

# calculate mean, median and mode for the specified columns
columns_to_analyze = ['mpg', 'cylinders', 'horsepower', 'weight']
for column in columns_to_analyze:
    mean_value = df[column].mean()
    median_value = df[column].median()
    mode_value = df[column].mode()[0]

    print(f"\nStatistics for {column}:")
    print(f"Mean: {mean_value}")
    print(f"Median: {median_value}")
    print(f"Mode: {mode_value}")

# plot histograms for the variables
plt.figure(figsize=(12, 8))
for i, col in enumerate(columns_to_analyze, start=1):
    plt.subplot(2, 2, i)
    sns.histplot(df[col], kde=True, bins=20, color='skyblue')
    plt.title(f"Histogram of {col}")
plt.tight_layout()
plt.show()

# plot boxplots for the variables
plt.figure(figsize=(12, 8))
for i, col in enumerate(columns_to_analyze, start=1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=df, x=col, color='lightgreen')
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()

# get skeweness value
skew_value = df.skew(axis=0, skipna=True, numeric_only=True)
print("\nSkewness values for numeric columns:")
print(skew_value)

# get kurtosis value
kurtosis_value = df.kurt(axis=0, skipna=True, numeric_only=True)
print("\nKurtosis values for numeric columns:")
print(kurtosis_value)

# check relationships between all columns
sns.pairplot(df[columns_to_analyze], diag_kind='kde', plot_kws={'alpha': 0.7})
plt.suptitle("Plot of mpg, horsepower, weight, and cylinders", y=1.02)
plt.show()

# correlation heatmap
correlation_matrix = df[columns_to_analyze].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap - Pearson")
plt.show()

correlation_matrix = df[columns_to_analyze].corr(method='spearman')
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap - Spearman")
plt.show()

# scatter plots for comparisions in pairs
plt.figure(figsize=(12, 8))
for i, col in enumerate(columns_to_analyze[1:], start=1):
    plt.subplot(2, 2, i)
    sns.scatterplot(data=df, x=col, y='mpg', alpha=0.7)
    plt.title(f"mpg vs {col}")
plt.tight_layout()
plt.show()

# generate scatter plots for all pairs of columns
plt.figure(figsize=(12, 8))
column_pairs = list(itertools.combinations(columns_to_analyze, 2))
for i, (x_col, y_col) in enumerate(column_pairs, start=1):
    plt.subplot(2, 3, i)
    sns.scatterplot(data=df, x=x_col, y=y_col, alpha=0.7, color='blue')
    plt.title(f"{x_col} vs {y_col}")
    plt.xlabel(x_col)
    plt.ylabel(y_col)

plt.tight_layout()
plt.show()

# shows the updated dataset
df
```

Univariate

The distribution of mpg appears right-skewed (positive skewness).
The presence of outliers can be seen in the boxplot.
The distribution doesn't appear symmetric, and the mean is probably greater than the median due to the skewness.

Null Hypothesis (H0): The mpg variable follows a normal distribution.
Alternative Hypothesis (H1): The mpg variable deviates from normality.

Bivariate

There appears to be a negative association between mpg and weight.
As the weight of a vehicle increases, the miles per gallon tend to decrease.
The relationship seems linear.

Null Hypothesis (H0): There is no association between mpg and weight.
Alternative Hypothesis (H1): There is an association between mpg and weight.
